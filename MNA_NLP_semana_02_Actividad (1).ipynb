*   **Pregunta 1.**

Busca y elimina todos los saltos de línea '\n' que se encuentran al final de cada comentario.

Una vez finalizado, imprime los primeros 10 comentarios del resultado obtenido.


```python
#Busca y elimina todos los saltos de línea '\n' que se encuentran al final de cada comentario.
docs = [re.sub(r'\n$', '', doc) for doc in docs]
```

```python
# Imprimir los primeros 10 comentarios
print(docs[0:10])
```

*   **Pregunta 2.**  

Busca e imprime todas las palabras que terminan con dos o más signos de admiración seguidos, por ejemplo "!!!".

Debes imprimir tanto la palabra como la totalidad de signos de admiración que le siguen.

Indica cuántos resultados obtuviste.



```python
matches = []
for doc in docs:
    found = re.findall(r'\b(\w+)(!{2,})', doc)
    if found:
        matches.extend(found)

print("Palabras encontradas que terminan con dos o más signos de admiración:")
for word, exclamations in matches:
    print(f"{word}{exclamations}")

print(f"\nTotal de resultados obtenidos: {len(matches)}")
```

*   **Pregunta 3.**  

Busca e imprime todas las palabras que están escritas totalmente en mayúsculas. Cada coincidencia debe ser una sola palabra.

Indica cuántas palabras encontraste.



```python
uppercase_words = []
for doc in docs:
    found_words = re.findall(r'\b[A-Z]+\b', doc)
    if found_words:
        uppercase_words.extend(found_words)

print("Palabras encontradas en mayúsculas:")
for word in uppercase_words:
    print(word)

print(f"\nTotal de palabras encontradas: {len(uppercase_words)}")
```

*   **Pregunta 4.**  

Busca e imprime los comentarios en donde todos los caracteres alfabéticos (letras) están en mayúsculas.

Cada coincidencia encontrada debe ser todo el comentario/enunciado.

Indica cuántos resultados obtuviste.


```python
all_caps_comments = []
for doc in docs:
    alphabetic_chars = "".join(filter(str.isalpha, doc))
    if alphabetic_chars and alphabetic_chars.isupper():
        all_caps_comments.append(doc)

print("Comentarios en donde todos los caracteres alfabéticos están en mayúsculas:")
for comment in all_caps_comments:
    print(comment)

print(f"\nTotal de resultados obtenidos: {len(all_caps_comments)}")
```

*   **Pregunta 5.**  

Busca e imprime todas las palabras que tengan una vocal acentuada, del tipo á, é, í, ó, ú.

Indica cuántos resultados obtuviste.

```python
accented_words = []
for doc in docs:
    found_accented_words = re.findall(r'\b[a-zA-Z]*[áéíóúÁÉÍÓÚ][a-zA-ZáéíóúÁÉÍÓÚ]*\b', doc)
    if found_accented_words:
        accented_words.extend(found_accented_words)

print("Palabras encontradas con vocales acentuadas:")
for word in accented_words:
    print(word)

print(f"\nTotal de palabras encontradas: {len(accented_words)}")
```

*   **Pregunta 6.**  

Busca e imprime todas las cantidades numéricas monetarias, enteras o con decimales, que inician con el símbolo $\$$.

Indica cuántos resultados obtuviste.

```python
monetary_amounts = []
for doc in docs:
    found_amounts = re.findall(r'\$[0-9]+(?:\.[0-9]{2})?', doc)
    if found_amounts:
        monetary_amounts.extend(found_amounts)

print("Cantidades monetarias encontradas:")
for amount in monetary_amounts:
    print(amount)

print(f"\nTotal de resultados obtenidos: {len(monetary_amounts)}")
```

*   **Pregunta 7.**  

Busca e imprime todas las palabras que sean variantes de la palabra "love", sin importar si incluyen mayúsculas o minúsculas, o la manera en que esté conjugada o alguna otra variación que se haga con dicha palabra.

Indica cuántos resultados obtuviste.

```python
love_variants = []
for doc in docs:
    found_variants = re.findall(r'\blov(?:e|ed|ing|es)?\b', doc, re.IGNORECASE)
    if found_variants:
        love_variants.extend(found_variants)

print("Variantes de 'love' encontradas:")
for variant in love_variants:
    print(variant)

print(f"\nTotal de resultados obtenidos: {len(love_variants)}")
```

*   **Pregunta 8.**  

Busca e imprime todas las palabras, variantes de "so" y "good", que tengan dos o más "o" en "so" y 3 o más "o" en good.

Indica cuántas encontraste.


```python
so_good_variants = []
for doc in docs:
    found_variants = re.findall(r'\bs(o{2,})\b|\bg(o{3,})d\b', doc, re.IGNORECASE)
    if found_variants:
        for variant_tuple in found_variants:
            for variant_part in variant_tuple:
                if variant_part:
                    so_good_variants.append(variant_part)

print("Variantes de 'so' y 'good' encontradas:")
for variant in so_good_variants:
    print(variant)

print(f"\nTotal de resultados obtenidos: {len(so_good_variants)}")
```

*   **Pregunta 9.**  

Busca e imprime todas las palabras que tengan una longitud mayor estrictamente a 10 caracteres alfabéticos.

No se consideran los signos de puntuación o caracteres especiales en la longitud de estas cadenas, solo caracteres alfabéticos en mayúsculas o minúsculas.

Indica la cantidad de palabras encontradas.


```python
long_alphabetic_words = []
for doc in docs:
    words_in_doc = re.findall(r'[a-zA-Z]+', doc)
    for word in words_in_doc:
        if len(word) > 10:
            long_alphabetic_words.append(word)

print("Palabras encontradas con longitud mayor a 10 caracteres alfabéticos:")
for word in long_alphabetic_words:
    print(word)

print(f"\nCantidad de palabras encontradas: {len(long_alphabetic_words)}")
```

*   **Pregunta 10.**  

Busca e imprime todas las palabras que inician con una letra mayúscula y terminan con una minúscula, pero que además no sea la primera palabra del comentario/string.

Indica la cantidad de resultados obtenidos.

```python
capitalized_not_first = []
for doc in docs:
    found = re.findall(r'.+?\b([A-Z][a-z]+)\b', doc)
    if found:
        capitalized_not_first.extend(found)

print("Palabras que inician con Mayúscula y terminan con minúscula (excluyendo la primera del comentario):")
for word in capitalized_not_first:
    print(word)

print(f"\nTotal de resultados obtenidos: {len(capitalized_not_first)}")
```

*   **Pregunta 11.**  

Busca e imprime la secuencia de dos o más palabras que están separadas por un guion, "-", sin que tengan espacios en blanco entre ellas.

Por ejemplo "Go-Kart" sería válido, pero "Go  -Kart" o "Go  -  Kart" no lo serían.

Indica la cantidad de resultados obtenidos.

```python
hyphenated_words = []
for doc in docs:
    found = re.findall(r'\b\w+(?:-\w+)+\b', doc)
    if found:
        hyphenated_words.extend(found)

print("Secuencias de palabras con guion encontradas:")
for sequence in hyphenated_words:
    print(sequence)

print(f"\nTotal de resultados obtenidos: {len(hyphenated_words)}")
```

*   **Pregunta 12.**  

Busca e imprime todas las palabras que terminan en "ing" o "ed".

Indica la cantidad de palabras que encontraste de cada una.

```python
words_ing = []
words_ed = []

for doc in docs:
    ing_found = re.findall(r'\b\w+ing\b', doc, re.IGNORECASE)
    words_ing.extend(ing_found)

    ed_found = re.findall(r'\b\w+ed\b', doc, re.IGNORECASE)
    words_ed.extend(ed_found)

print(f"Palabras que terminan en 'ing': {words_ing[:10]}... (se muestran las primeras 10)")
print(f"Palabras que terminan en 'ed': {words_ed[:10]}... (se muestran las primeras 10)")

print(f"\nCantidad de palabras terminadas en 'ing': {len(words_ing)}")
print(f"Cantidad de palabras terminadas en 'ed': {len(words_ed)}")
```

#**Parte 3. Proceso de limpieza.**

*   **Pregunta 13.**  

Ahora realiza un proceso de limpieza del corpus que incluya los siguientes procesos:

*   Solo se deben considerar caracteres alfabéticos. Es decir, se eliminan todos los signos de puntuación y caracteres especiales.
*   Todos los caracteres alfabéticos se transforman a minúsculas.
*   Se deben eliminar todos los espacios en blanco adicionales que se puedan encontrar en cada comentario.

Al finalizar dicho proceso de limpieza, imprime el resultado de los primeros 10 comentarios resultantes.
   




```python
cleaned_docs = []

for doc in docs:
    clean_text = re.sub(r'[^a-zA-Z]', ' ', doc).lower()

    clean_text = re.sub(r'\s+', ' ', clean_text).strip()

    cleaned_docs.append(clean_text)

print("Primeros 10 comentarios limpios:")
for i, doc in enumerate(cleaned_docs[:10]):
    print(f"{i+1}: {doc}")
```

*   **Pregunta 14.**  

Con el resultado de la limpieza obtenido en la pregunta anterior, realiza ahora un proceso de tokenización por palabras del corpus.

Es decir, al final de este proceso de tokenización, debes tener como resultado una lista de listas, donde cada comentario estará tokenizado por palabras.

Al terminar calcula el total de tokens obtenido en todo el corpus.

```python
tokenized_docs = [doc.split() for doc in cleaned_docs]

total_tokens = sum(len(tokens) for tokens in tokenized_docs)

print(f"Primer comentario tokenizado: {tokenized_docs[0]}")
print(f"Total de tokens en el corpus: {total_tokens}")
```
